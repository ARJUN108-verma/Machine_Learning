K means Clustering:-

K-means clustering is an unsupervised machine learning algorithm used to partition data points into distinct groups, or clusters, based on similarity. It aims to create clusters where data points within a cluster are more similar to each other than to those in other clusters.

```python
import seaborn as sns
```

```python
df  = sns.load_dataset('iris')
```

```python
df.head()
```

```python
x = df[['sepal_length','sepal_width','petal_length','petal_width',]]
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
```

```python
x_scaled = sc.fit_transform(x)
```

```python
sns.scatterplot(x = x['sepal_length'], y = x['sepal_width'], hue= df['species'])
```

```python
from sklearn.cluster import KMeans
```

```python
kmean = KMeans(n_clusters=3, random_state=42)
```

```python
df['cluster'] = kmean.fit_predict(x_scaled)
```

```python
df.head()
```

```python
sns.scatterplot(x = x['sepal_length'], y = x['sepal_width'], hue= df['cluster'])
```

```python
centre = kmean.cluster_centers_
```

```python
centroids = sc.inverse_transform(centre)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize= (8,6))
sns.scatterplot(x = df['sepal_length'], y = df['petal_width'], hue= df['cluster'], palette='Set1', s = 60)
plt.scatter(x = centroids[:,0],y = centroids[:, 1], color = 'black', s = 200, label = 'Centroids')
```
