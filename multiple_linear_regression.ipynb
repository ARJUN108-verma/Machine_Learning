```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
```

```python
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.linear_model import LogisticRegression

# Define columns
categorical_cols = ['mainroad', 'guestroom', 'basement', 'airconditioning', 'furnishingstatus']
numerical_cols = ['area', 'bedrooms', 'bathrooms']  # change this based on your dataset

# Preprocessor to handle both types
preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), numerical_cols),
        ('cat', OneHotEncoder(drop='first'), categorical_cols)
    ]
)

# Complete pipeline
pipeline = Pipeline([
    ('preprocessing', preprocessor),
    ('classifier', LogisticRegression())
])

# Fit the pipeline
pipeline.fit(x_train, y_train)
```

```python
a = pipeline.score(x_train, y_train)
a
```

```python
b = pipeline.score(x_test, y_test)
b
```
