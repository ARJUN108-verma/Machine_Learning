```python
from sklearn.datasets import load_wine
```

```python
df = load_wine()
```

```python
print(df.DESCR)
```

```python
x = df.data
y = df.target
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
```

```python
s_scaled = sc.fit_transform(x)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(s_scaled,y,test_size=0.2, random_state=42)
```

```python
from xgboost import XGBClassifier
```

```python
model = XGBClassifier()
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_test, y_test)
```
